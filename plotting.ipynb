```python
# (1) Read in the file cisco_stock.csv.  We're interestd in the open and
#     close rates for this period -- column indexes 1 and 5 in the file.

a = np.loadtxt('cisco_stock.csv', delimiter=',', usecols=[1,5])
a
```

```python
# (2) Show a scatter plot of open vs. close prices.

opening_prices = a[:, 0]
closing_prices = a[:, 1]

plt.plot(opening_prices, closing_prices, 'b+', label='open vs. close')
plt.legend()
plt.show()
```

```python
# (3) Show a line plot of closing prices.

plt.plot(np.arange(closing_prices.size), closing_prices)
plt.show()
```

```python
# (4) Show a line plot of the difference between opening and closing
#     prices for this period.

difference = closing_prices - opening_prices
difference
```

```python
plt.plot(np.arange(difference.size), difference, label='diff')
plt.plot(np.arange(closing_prices.size), closing_prices, label='closing')
plt.legend()
plt.show()
```
